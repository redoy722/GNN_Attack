**Extracting Orbit based nodes list**

```python
orbits = {}
for orbit_type, group in df_2d.groupby('Orbit_type'):
    orbits['orbit_'+ str(orbit_type)] = list(group['node_number'])
```

```python
for key, value in orbits.items():
    print(key + ' = ' + str(value))
```

**Finding the RL states based on the orbit edge pairs**

```python
import itertools

checkList=graph_ed.tolist()

orbit_pairs = {}
compared_orbits = {}

for key, values in orbits.items():
    pair_list = []
    for i in range(len(values)):
        for j in range(i, len(values)):
            pair_list.append([values[i], values[j]])
            if i != j:
                pair_list.append([values[j], values[i]])
    orbit_num = key.split("_")[-1]
    orbit_pair_key = f"orbit_{orbit_num}{orbit_num}"
    orbit_pairs[orbit_pair_key] = pair_list


for orbit1, orbit2 in itertools.product(orbits.keys(), repeat=2):
    if orbit1 < orbit2:
        pair_list = []
        for value1 in orbits[orbit1]:
            for value2 in orbits[orbit2]:
                pair_list.append([value1, value2])
        orbit_nums = orbit1.split("_")[-1] + orbit2.split("_")[-1]
        orbit_pair_key = f"orbit_{orbit_nums}"
        orbit_pairs[orbit_pair_key] = pair_list


for orbit_pair_key, orbit_pair_list in orbit_pairs.items():
    compared_pair_list = []
    for pair in orbit_pair_list:
        if pair in checkList:
            compared_pair_list.append(pair)
        #else:
            #compared_pair_list.append([])

    compared_orbit_key = f"compared_{orbit_pair_key}"
    compared_orbits[compared_orbit_key] = compared_pair_list
```
